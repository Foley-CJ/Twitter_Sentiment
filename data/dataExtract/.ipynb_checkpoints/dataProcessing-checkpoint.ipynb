```python
import tweepy
import pandas as pd
import time
from dataGenConstants import companies, finaceMasterUrl
from nltk.sentiment.vader import SentimentIntensityAnalyzer 
```

```python
# Attach new sentiment

class TwitterSiaCustom(SentimentIntensityAnalyzer):
    
    def __init__(self,consumer_key, consumer_secret, access_token, access_token_secret):
        SentimentIntensityAnalyzer.__init__(self)

        # OAuth process, using the keys and tokens
        auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
        auth.set_access_token(access_token, access_token_secret)

        # Creation of the actual interface, using authentication
        self.api = tweepy.API(auth)
        
    def BaseSentiment(self,sentence):
        cmpndScore = self.polarity_scores(sentence)['compound']
        if (cmpndScore >= 0.4):
            return 'positive'
        elif (cmpndScore <= -0.4):
            return 'negative'
        else:
            return 'neutral' 
        
    def extractDataSentiment(self, twitterHandle, companyName, pages, sentiment=True):
        zed = 0
        with open('raw/twitter/'+companyName+'_twitter.txt','w') as f:
            for page_number in range(0,pages):
                print("##############"+str(page_number)+"#############")
                for status in tweepy.Cursor(self.api.user_timeline, screen_name=twitterHandle,page=page_number, tweet_mode='extended').items():
                    zed = zed + 1
                    print(zed)
  
                    #if zed >= 10:
                    #    break
                    if str(status._json['in_reply_to_user_id_str']) != 'None':
                        continue
                    if sentiment:
                        text = status._json['full_text']
                        crt_date = time.strftime('%Y-%m-%d', time.strptime(status._json['created_at'],'%a %b %d %H:%M:%S +0000 %Y'))
                        dataLine = crt_date+","+self.BaseSentiment(text)+"\n"
                        print(dataLine)
                        f.write(dataLine)
                    else:
                        data = ','.join(['"'+str(status._json[x].encode('ascii','ignore')).replace('/n',' ')+'"' for x in ['text', 'created_at']])
                        print(data)
                        
    def aggregateData(self, companyName):
        rawData = pd.read_csv('raw/twitter/'+companyName+'_twitter.txt',names=['date','sentiment'],header=None)
        aggData = rawData.groupby(rawData.columns.tolist()).size().reset_index().rename(columns={0:'responses'})
        aggData.to_csv('agg/twitter/'+companyName+'_twitter.txt',index=False)
        

        
```

```python
tcon = TwitterSiaCustom(consumer_key, consumer_secret, access_token, access_token_secret)
tcon.extractDataSentiment('@NortonOnline','norton_antivirus',10,True)
tcon.aggregateData('norton_antivirus')
```
